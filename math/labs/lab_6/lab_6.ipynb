```python
import numpy as np
import scipy
import sympy
import matplotlib.pyplot as plt
```

```python
# Функция №1
def f1(a, b, step):
    x = np.arange(a, b, step)
    y = (1 + x ** 2) * np.exp(-x ** 2)
    return x, y
plt.figure(figsize = (9, 3)) # создаем полотно
x, y = f1(-10, 10, 0.05) # разбиваем отрезок на части
plt.plot(x, y, lw = 1, c = 'b', ls = '-') # выводим график
```
