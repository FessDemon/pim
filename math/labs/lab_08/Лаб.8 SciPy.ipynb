<h3><b>Задание 1. Символьное дифференцирование функции одной переменной</b></h3>

```python
from sympy import *

init_printing()
x = Symbol('x')

def f(x):
    return cos(x) * log(tan(x)) - log(tan(x / 2))
```

```python
print('Исходная функция:')
f(x)
```

```python
df = diff(f(x),x)
print('Производная исходной функции')
df
```

<h3><b>Задание 2. Частные производные</b></h3>

```python
x = Symbol('x')
y = Symbol('y')

def f(x, y):
    return 2 * exp(sin(x * y))

print('Исходная функция:')
f(x, y)
```

```python
df = diff(f(x, y), x)
print('Первая производная по x:')
df
```

```python
df2 = diff(f(x, y), x, 2)
print('Вторая производная по х:')
df2
```

```python
dfx = diff(f(x, y), x)
#print('df/dx:')
#dfx
dfxy = diff(dfx, y)
print('d(df/dx)/dy:')
dfxy
```

```python
dfy = diff(f(x, y), y)
#print('df/dy:')
#dfy
dfyx = diff(dfy, x)
print('d(df/dy)/dx:')
dfyx
```

```python
if dfxy == dfyx:
    print('Теорема о равенстве смешанных производных выполняется')
```

<h3><b>Задание 3. Поиск минимума функции одной переменной.</b></h3>

```python
from scipy.optimize import minimize
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return log(x)/sqrt(x)

xArr = np.arange(0.01, 5., 0.05) # Диапазон изменения x
yArr = np.array([f(x) for x in xArr])
plt.plot(xArr, yArr)
plt.grid(True)
plt.axis([0.01, 5, -15, 5])
plt.show()
```

```python
import numpy as np

def f(x):
    return -(log(x[0]) / sqrt(x[0]))

x0 = np.array([0.5, 1, 1.5, 2, 2.5, 5])

for i in range(len(x0)):
    resFunc2Metod1 = minimize(f, x0[i]) # В качестве второго параметра указывается начальная точка
    print("Minimized F(x) (standard method) при начальной точке x0 = ", x0[i], ": x = " , resFunc2Metod1.x)
    print("Number of iterations: ", resFunc2Metod1.nit)

for i in range(len(x0)):
    resFunc2Metod2 = minimize(f, x0[i], method = 'BFGS')
    print("Minimized F(x) (BFGS method) при начальной точке x0 = ", x0[i], ": x = " , resFunc2Metod2.x)
    print("Number of iterations: ", resFunc2Metod2.nit)
```

<h3><b>Задание 4. Поиск минимума функции многих переменных</b></h3>

```python
def target_func(x):
    return x[0] ** 2 + x[1] ** 2 - 4 * (x[0] - x[1])

x0 = np.array([0, 0])

print('Минимизация методом Нелдера-Мида:')
res1 = minimize(target_func, x0, method = 'nelder-mead')
print("Minimized F(x) (nelder-mead method): ", res1.fun, " for x = ", res1.x)
print("Number of iterations: ", res1.nit, "\n")

print('Минимизация методом сопряженных градиентов:')
res2 = minimize(target_func, x0, method = 'CG')
print("Minimized F(x) (CG method): ", res2.fun, " for x = ", res2.x)
print("Number of iterations: ", res2.nit, "\n")

print('Минимизация методом BFGS:')
res3 = minimize(target_func, x0, method = 'BFGS')
print("Minimized F(x) (BFGS method): ", res3.fun, " for x = ", res3.x)
print("Number of iterations: ", res3.nit, "\n")
```

Вывод: Самый быстрый - метод сопряженных градиентов.

```python
print("Абсолютные погрешности методов оптимизации:\n")
print("Абсолютная погрешность метода nelder-mead: ", abs(res1.fun - (-8)), "\n")

print("Абсолютная погрешность метода сопряженных градиентов (CG): ", abs(res2.fun - (-8)), "\n")

print("Абсолютная погрешность метода BFGS: ", abs(res3.fun - (-8)), "\n")
```

<h3><b>Задание 5. Метод наименьших квадратов.</b></h3>

```python
from numpy.polynomial import Polynomial
import numpy as np
import matplotlib.pyplot as plt
    
x = np.array([0.5, 0.75, 1.00, 1.25, 1.50, 1.75])
y = np.array([5.10, 3.50, 3.10, 2.50, 2.30, 2.00])
xh = np.linspace(x.min(), x.max(), 100) # точки (равномерно распределенные) для построения графиков аппроксимаций
#degrees = np.array([1, 2]) # степени аппроксимирующих многочленов

approximation1 = Polynomial.fit(x, y, 1) # нахождение коэффициентов аппроксимирующего многочлена 1-ой степени
approximation2 = Polynomial.fit(x, y, 2) # нахождение коэффициентов аппроксимирующего многочлена 2-ой степени

yy1 = approximation1(x) # нахождение вектора значений аппроксимирующего многочлена 1-ой степени в заданных (исходных) точках
SKO1 = np.sum((y - yy1) ** 2) # нахождение суммы квадратов отклонений
yy2 = approximation2(x) # нахождение вектора значений аппроксимирующего многочлена 2-ой степени в заданных (исходных) точках
SKO2 = np.sum((y - yy2) ** 2) # нахождение суммы квадратов отклонений
# построение графиков
plt.figure(figsize = (16, 8))
plt.title("Полиномиальная аппроксимация") # вывод заголовка
plt.xlabel("$x$") # подпись оси x
plt.ylabel("$y$") # подпись оси y
plt.scatter(x, y, marker = "x", color = "red", label = "Исходные данные") # построение точечного графика для исходных данных
plt.plot(xh, approximation1(xh), color = "green", label = "Аппроксимация полиномом 1-ой степени\n(сумма квадратов отклонений = " + str(SKO1) + ")") # построение графика аппроксимирующей функции
plt.plot(xh, approximation2(xh), color = "blue", label = "Аппроксимация полиномом 2-ой степени\n(сумма квадратов отклонений = " + str(SKO2) + ")") # построение графика аппроксимирующей функции
plt.legend() # вывод легенды
```

```python
# Задание 5.6

from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
    
x = np.array([0.5, 0.75, 1.00, 1.25, 1.50, 1.75])
y = np.array([5.10, 3.50, 3.10, 2.50, 2.30, 2.00])
xh = np.linspace(x.min(), x.max(), 100) # точки (равномерно распределенные) для построения графиков аппроксимаций
```

```python
# Задание 5.6. Функция №3

def appr_f3(x, a, m):
    return a * (x ** m)

try:
    popt, pcov = optimize.curve_fit(appr_f3, x, y, p0 = [0, 0])
    yy = appr_f3(x, popt[0], popt[1]) # нахождение вектора значений 3-ей аппроксимирующей функции в заданных (исходных) точках
    SKO = np.sum((y - yy) ** 2) # нахождение суммы квадратов отклонений

    # построение графиков
    plt.figure(figsize = (16, 8))
    plt.title("Аппроксимация") # вывод заголовка
    plt.xlabel("$x$") # подпись оси x
    plt.ylabel("$y$") # подпись оси y
    plt.scatter(x, y, marker = "x", color = "red", label = "Исходные данные") # построение точечного графика для исходных данных
    plt.plot(xh, appr_f3(xh, popt[0], popt[1]), color = "green", label = "Аппроксимация функцией №3 (a = " + str(popt[0]) + " m = " + str(popt[1]) + 
             ")\n(сумма квадратов отклонений = " + str(SKO) + ")") # построение графика аппроксимирующей функции
    plt.legend() # вывод легенды
except:
    print("Аппроксимация прошла неудачно")
```
