```python
import torch
import torch.nn as nn
```

```python
input_data = [[[
    [1., 1., 2., 4.,],
    [5., 6., 7., 8.],
    [3., 2., 1., 0.],
    [1., 2., 3., 4.]
]]]
input_data = torch.tensor(input_data)

print(f'input_data.shape = {input_data.shape}')
print(f'\ninput_data:\n{input_data}')
```

```python
max_pooling = nn.MaxPool2d(kernel_size=2, stride=2, return_indices=False, ceil_mode=False)
max_pooling
```

```python
output = max_pooling(input_data)
output
```

```python
max_pooling = nn.MaxPool2d(kernel_size=2, stride=2, return_indices=True, ceil_mode=False)
max_pooling
```

```python
output, indices = max_pooling(input_data)
print(f'output:\n{output}')
print(f'\nindices:\n{indices}')
```
