```python
import numpy as np

def get_weight(X, y):
    # Добавляем столбец единиц к матрице X
    X_b = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
    # Вычисляем веса с помощью нормального уравнения
    w = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
    return w
```

```python
import matplotlib.pyplot as plt

# Пример данных
X = np.array([[2], [3], [5]])  # Матрица признаков (один столбец)
y = np.array([1, 3, 4])            # Вектор целевых значений

# Получаем веса
weights = get_weight(X, y)

# Выводим веса
print("Вектор весов:", weights)

# Предсказания на основе полученных весов
X_b = np.concatenate((np.ones((X.shape[0], 1)), X), axis=1)
y_pred = X_b @ weights

# Визуализация данных и линии регрессии
plt.scatter(X, y, color='blue', label='Исходные данные')
plt.plot(X, y_pred, color='red', label='Линия регрессии')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Линейная регрессия')
plt.grid(True)
plt.legend()
plt.show()
```
